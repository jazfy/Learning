# 2. 抓取网页

```python
import requests
import re #导入解析库
```

```python
headers = {
    'user-agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_0) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3534.4 Safari/537.36'
} #添加参数
```

```python
r = requests.get('https://www.zhihu.com/explore',headers=headers)
pattern = re.compile('explore-feed.*?question_link.*?>(.*?)</a>',re.S) #正则
titles = re.findall(pattern,r.text) #正则解析
print(titles)
```

# 3. 抓取二进制数据

```python
import requests
```

```python
r = requests.get('https://github.com/favicon.ico')
```

```python
print(r.content) #图片等返回结果用 content 而非 text
```

## 3.1 保存图片

```python
r = requests.get('https://github.com/favicon.ico')
with open('favicon.ico','wb') as f:
    f.write(r.content)
    f.close()
```
